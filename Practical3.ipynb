```python
''' Problem 2
Grocery Shopping list
A user maintains a grocery list for monthly shopping.
Tasks:
-Create a grocery list
-Add new items to the list
-Remove purchased items
-Display the final shopping list'''

grocerylist = ["Eggs", "Candy", "Bread", "Cheese", "Strawberry", "Milk"]
newlist = ["Ice Cream", "Olive Oil"]
grocerylist.extend(newlist)
grocerylist.remove("Eggs")
print(grocerylist)
```

```python
'''Problem 3
A Person stores phone contact in a list
Create list of contacts
Add a new contact
Delete existing contact
Display contacts alphabetically'''

contacts = ["Asmi", "Mano", "Arshiya", "Pushkar"]
newcontacts = ["Anvi", "Neil", "Prisha"]
contacts.extend(newcontacts)
contacts.remove("Neil")
contacts.sort()
contacts
```

```python
'''Problem 4
Temperature Analysis:
Daily Temperature readings of a city are stored in a list for 10 days.
Tasks-
-Display temperatures of first and last 5 days
-Find highest and lowest temperatures
-Calculate average temperature'''

weatherreport = [25,24,19,27,18,20,23,30,26,29]
print("First 5 days report:",weatherreport[:6])
print("Last 5 days:",weatherreport[:-5])
print("Highest Temperature:",max(weatherreport))
print("Lowest Temperature:",min(weatherreport))
print("Average Temperature",sum(weatherreport)/len(weatherreport))
```

```python
'''Problem 5
Daily Expense Tracker
Tasks:
-Store daily expenses
-Display total and average expenses
-Identify highest spending day'''

dailyspend=[100,40,78,90,26]
print("Total Expenses-",sum(dailyspend))
print("Highest Spending Day-",max(dailyspend))
print("Lowest Spending Day-",min(dailyspend))
print("Average Expenses-",sum(dailyspend)/len(dailyspend))
```
